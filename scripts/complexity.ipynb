```python
import requests
import json
import matplotlib.pyplot as plt
import numpy as np
from types import SimpleNamespace
import time
```

```python
timeMess = []
n = []

def getBody():
    return {
        "grammar": {
            "terminals": [
            "(", ")", "ε"
            ],
            "nonTerminals": [
            "S"
            ],
            "productions": [
            {
                "leftSide": "S",
                "rightSide": "ε"
            },
            {
                "leftSide": "S",
                "rightSide": "(S)"
            },
            {
                "leftSide": "S",
                "rightSide": "SS"
            }
            ],
            "startSymbol": "S"
        },
        "derivationTree": {
            "content": "S",
            "classification": "NON_TERMINAL",
            "children": [
                {
                "content": "(",
                "classification": "TERMINAL"
                },
                {
                "content": "S",
                "classification": "NON_TERMINAL",
                "children": [
                    {
                    "content": "ε",
                    "classification": "TERMINAL"
                    },
                ]
                },
                {
                "content": ")",
                "classification": "TERMINAL"
                }
            ]
        },
        "word": "()"
}
```

```python
def validate_derivation_tree (payload):    
    headers = {'Content-type': 'application/json'}
    response = requests.post("https://studienarbeit.mohjohfox.de/api/validate/tree", headers=headers, data=json.dumps(payload))
    if response.status_code == 200:
        # takes response as json
        return response.json()     
    
def expand_derivation_tree(derivation_tree, depth):
    if depth <= 0:
        return derivation_tree
    
    new_tree = {
        "content": "S",
        "classification": "NON_TERMINAL",
        "children": [
            {
                "content": "(",
                "classification": "TERMINAL"
            },
            {
                "content": "S",
                "classification": "NON_TERMINAL",
                "children": [
                    {
                        "content": "ε",
                        "classification": "TERMINAL"
                    }
                ]
            },
            {
                "content": ")",
                "classification": "TERMINAL"
            }
        ]
    }
    
    if "children" in derivation_tree:
        derivation_tree["children"] = [expand_derivation_tree(new_tree, depth - 1)]
    
    return derivation_tree

def extend_derivation_tree(tree, depth):
    tree["derivationTree"] = expand_derivation_tree(tree["derivationTree"], depth)
    return tree
```

```python
start = 0
end = 0
amount_of_validations = 8

for i in range(amount_of_validations):
    tree = getBody()
    body = extend_derivation_tree(tree, i)
    print(str(i))
    start = time.time()
    result = validate_derivation_tree(body)
    end = time.time()
    print(result)
    
    n.append(i)
    timeMess.append(end - start)
```
