```python
def validate_derivation_tree (payload):    
    headers = {'Content-type': 'application/json'}
    #response = requests.post("https://studienarbeit.mohjohfox.de/api/validate/tree", headers=headers, data=json.dumps(payload))
    response = requests.post("http://localhost:8080/api/validate/tree", headers=headers, data=json.dumps(payload))
    if response.status_code == 200:
        # takes response as json
        return response.json()     
    
def find_last_non_terminal_node(tree):
    if "children" not in tree:
        return None
    
    last_child = tree["children"][-1]
    if last_child["classification"] == "NON_TERMINAL":
        return last_child
    
    return find_last_non_terminal_node(last_child)

def expand_derivation_tree(tree):
    new_step = {
        "content": "S",
        "classification": "NON_TERMINAL",
        "children": [
            {
                "content": "epsilon",
                "classification": "TERMINAL"
            }
        ]
    }
    
    last_non_terminal = find_last_non_terminal_node(tree)
    
    if last_non_terminal is not None:
        if "children" not in last_non_terminal:
            last_non_terminal["children"] = []
        last_non_terminal["children"].append(new_step)
    
    return tree

# Original derivation tree in JSON format
original_tree_json = '''
{
    "content": "S",
    "classification": "NON_TERMINAL",
    "children": [
        {
            "content": "S",
            "classification": "NON_TERMINAL",
            "children": [
                {
                    "content": "epsilon",
                    "classification": "TERMINAL"
                },
                {
                    "content": "S",
                    "classification": "NON_TERMINAL",
                    "children": [
                        {
                            "content": "epsilon",
                            "classification": "TERMINAL"
                        }
                    ]
                }
            ]
        }
    ]
}
'''
```

```python
start = 0
end = 0
amount_of_validations = 60
tree = json.loads(original_tree_json)
timeMess = []
n = []

for i in range(amount_of_validations):
    # Expand the derivation tree
    expanded_tree = expand_derivation_tree(tree)

    # Convert the expanded tree back to JSON
    expanded_tree_json = json.dumps(expanded_tree, indent=4)
    tree = json.loads(expanded_tree_json)

    for k in range(100):
        body = getBody()
        body["derivationTree"] = tree

        start = time.time()
        result = validate_derivation_tree(body)
        end = time.time()
        
        if (k == 50):
            n.append(i)
            timeMess.append(end - start)
```

```python
# Comment out the next three lines for default sizing
f = plt.figure()
f.set_figwidth(20)
f.set_figheight(8)
plt.plot(n, timeMess)
plt.xlabel('n')
plt.ylabel('Zeit')
plt.show()
```
